```python
import pandas as pd
import numpy as np
from pathlib import Path

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    classification_report,
    accuracy_score,
    confusion_matrix,
    roc_curve,
    auc,
    precision_recall_curve,
    average_precision_score
)
```

```python
sns.set(style="whitegrid", context="talk")
plt.rcParams["figure.figsize"] = (12, 6)
```

```python
ticker = "AAPL"
path = Path("../data/processed") / f"{ticker}_features.csv"
df = pd.read_csv(path, index_col=0, parse_dates=True)
```

```python
print(f"Rows: {len(df)}, Columns: {len(df.columns)}")
df.tail(3)
```

```python
display(df.describe().T)
```

```python
missing_pct = df.isna().mean().sort_values(ascending=False)
ax = missing_pct.plot(kind="bar", color="#cc444b", alpha=0.8)
ax.set_title("Missing value percentage by column")
ax.set_ylabel("Fraction missing")
plt.tight_layout()
plt.show()
```

```python
ax = df["Close"].plot(title=f"{ticker} Close Price Over Time", color="#1f77b4")
ax.set_xlabel("Date")
ax.set_ylabel("Price")
plt.tight_layout()
plt.show()
```

```python
df["Daily_Return"] = df["Close"].pct_change()
sns.histplot(df["Daily_Return"].dropna(), bins=50, kde=True, color="#2ca02c")
plt.title("Distribution of daily returns")
plt.xlabel("Daily return")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
features_core = ["SMA_20", "EMA_50", "RSI_14", "BB_upper", "BB_lower", "Daily_Return", "Lag_1"]
available_features = [f for f in features_core if f in df.columns]
```

```python
fig, axes = plt.subplots(nrows=int(np.ceil(len(available_features)/3)), ncols=3, figsize=(16, 10))
axes = axes.flatten()
for i, f in enumerate(available_features):
    sns.histplot(df[f].dropna(), bins=50, kde=True, ax=axes[i], color="#9467bd")
    axes[i].set_title(f)
for j in range(i+1, len(axes)):
    axes[j].axis("off")
plt.suptitle("Feature value distributions")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.boxplot(data=df[available_features], orient="h")
plt.title("Feature boxplots (outliers and spread)")
plt.tight_layout()
plt.show()
```

```python
plot_cols = ["Close"] + [c for c in ["SMA_20", "EMA_50"] if c in df.columns]
df[plot_cols].plot(title="Price with SMA/EMA overlays", linewidth=1.2)
plt.xlabel("Date")
plt.ylabel("Price")
plt.tight_layout()
plt.show()
```

```python
if "RSI_14" in df.columns:
    df["RSI_14"].plot(color="#ff7f0e", title="RSI (14) over time")
    plt.axhline(70, color="red", linestyle="--", alpha=0.7, label="Overbought (70)")
    plt.axhline(30, color="green", linestyle="--", alpha=0.7, label="Oversold (30)")
    plt.legend()
    plt.tight_layout()
    plt.show()

bb_cols = [c for c in ["Close", "BB_upper", "BB_lower"] if c in df.columns]
if len(bb_cols) == 3:
    ax = df[["Close"]].plot(title="Bollinger Bands and Price", color="#1f77b4", linewidth=1.2)
    df[["BB_upper", "BB_lower"]].plot(ax=ax, color=["#d62728", "#2ca02c"], linewidth=0.8)
    ax.fill_between(df.index, df["BB_lower"], df["BB_upper"], color="#c5d5ea", alpha=0.3)
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.tight_layout()
    plt.show()
```

```python
# our goal is to predict if AAPL return tomorrow is going to be positive
df["Target"] = (df["Daily_Return"].shift(-1) > 0).astype(int)

#clean
model_cols = available_features + ["Target", "Close"]
df_model = df[model_cols].dropna()

X = df_model[available_features]
y = df_model["Target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size=0.2)

print(f"Train size: {len(X_train)}, Test size: {len(X_test)}")
```

```python
model = RandomForestClassifier(n_estimators=300, max_depth=None, random_state=42, n_jobs=-1)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()

fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, color="#1f77b4", label=f"ROC curve (AUC = {roc_auc:.3f})")
plt.plot([0, 1], [0, 1], color="gray", linestyle="--")
plt.title("ROC Curve")
plt.xlabel("False positive rate")
plt.ylabel("True positive rate")
plt.legend(loc="lower right")
plt.tight_layout()
plt.show()

precision, recall, thresholds = precision_recall_curve(y_test, y_prob)
ap = average_precision_score(y_test, y_prob)
plt.plot(recall, precision, color="#ff7f0e", label=f"PR curve (AP = {ap:.3f})")
plt.title("PrecisionRecall Curve")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.legend(loc="lower left")
plt.tight_layout()
plt.show()
```

```python
importances = pd.Series(model.feature_importances_, index=available_features).sort_values(ascending=True)

plt.barh(importances.index, importances.values, color="#6baed6")
plt.title("Random Forest Feature Importances")
plt.xlabel("Importance")
plt.tight_layout()
plt.show()

if "RSI_14" in X_test.columns:
    temp = pd.DataFrame({"RSI_14": X_test["RSI_14"], "Pred_Prob": y_prob})
    sns.scatterplot(data=temp, x="RSI_14", y="Pred_Prob", alpha=0.5)
    plt.title("Predicted probability vs RSI")
    plt.xlabel("RSI_14")
    plt.ylabel("P(up tomorrow)")
    plt.tight_layout()
    plt.show()
```

```python
test_index = X_test.index
df_plot = df.loc[test_index, ["Close"]].copy()
df_plot["Pred_Up"] = y_pred
df_plot["Pred_Prob"] = y_prob

plt.plot(df_plot.index, df_plot["Close"], label="Close", color="#1f77b4")
up_idx = df_plot.index[df_plot["Pred_Up"] == 1]
down_idx = df_plot.index[df_plot["Pred_Up"] == 0]
plt.scatter(up_idx, df_plot.loc[up_idx, "Close"], label="Predicted Up", color="#2ca02c", marker="^", s=60)
plt.scatter(down_idx, df_plot.loc[down_idx, "Close"], label="Predicted Down", color="#d62728", marker="v", s=60)
plt.title(f"{ticker} price with predicted signals")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.tight_layout()
plt.show()

sns.lineplot(x=df_plot.index, y=df_plot["Pred_Prob"], color="#9467bd")
plt.title("Predicted probability of up-move over time")
plt.xlabel("Date")
plt.ylabel("P(up)")
plt.tight_layout()
plt.show()
```

```python
df_bt = df.loc[test_index, ["Close", "Daily_Return"]].copy()
df_bt["Signal"] = y_pred 
df_bt["Strategy_Return"] = df_bt["Daily_Return"] * df_bt["Signal"].shift(1)  # enter next day

df_bt["Cum_Market"] = (1 + df_bt["Daily_Return"]).cumprod()
df_bt["Cum_Strategy"] = (1 + df_bt["Strategy_Return"].fillna(0)).cumprod()

plt.plot(df_bt.index, df_bt["Cum_Market"], label="Market (Close)", color="#1f77b4")
plt.plot(df_bt.index, df_bt["Cum_Strategy"], label="Strategy (Pred up)", color="#ff7f0e")
plt.title("Cumulative returns: market vs strategy")
plt.xlabel("Date")
plt.ylabel("Growth of $1")
plt.legend()
plt.tight_layout()
plt.show()

# Distribution of strategy returns
sns.histplot(df_bt["Strategy_Return"].dropna(), bins=50, kde=True, color="#ff9896")
plt.title("Distribution of strategy daily returns")
plt.xlabel("Daily return")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
import joblib

model_path = Path("../outputs/models") / f"{ticker}_rf_model.joblib"
joblib.dump(model, model_path)

# Save predictions
pred_path = Path("../outputs") / f"{ticker}_test_predictions.csv"
pd.DataFrame({
    "Date": test_index,
    "Pred_Class": y_pred,
    "Pred_Prob": y_prob
}).to_csv(pred_path, index=False)

print(f"Saved model to: {model_path}")
print(f"Saved predictions to: {pred_path}")
```
