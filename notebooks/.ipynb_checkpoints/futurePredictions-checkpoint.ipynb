```python
import pandas as pd
import numpy as np
import yfinance as yf
import joblib
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", context="talk")
plt.rcParams["figure.figsize"] = (12, 6)

ticker = "AAPL"
model_path = Path("../outputs/models") / f"{ticker}_rf_model.joblib"

#load the model!!!
model = joblib.load(model_path)
print("Model loaded:", type(model))
print("Model expects features:", list(model.feature_names_in_))
```

```python
df_future = yf.download(ticker, start="2025-01-01", end="2025-12-31")
print(f"Rows: {len(df_future)}, Columns: {list(df_future.columns)}")
df_future.head()
```

```python
df_future["SMA_20"] = df_future["Close"].rolling(20).mean()
df_future["EMA_50"] = df_future["Close"].ewm(span=50, adjust=False).mean()

df_future["Daily_Return"] = df_future["Close"].pct_change()
df_future["Lag_1"] = df_future["Close"].shift(1)

df_future[["Close","SMA_20","EMA_50","Daily_Return","Lag_1"]].tail()
```

```python
features = list(model.feature_names_in_) 
X_future = df_future[features].dropna()

df_future = df_future.loc[X_future.index]

# Predictions
df_future["Pred_Class"] = model.predict(X_future)
df_future["Pred_Prob"] = model.predict_proba(X_future)[:, 1]

df_future[["Close","Pred_Class","Pred_Prob"]].head()
```

```python
plt.figure(figsize=(14,6))
plt.plot(df_future.index, df_future["Close"], label="Close Price", color="blue")

up_days = df_future[df_future["Pred_Class"] == 1].index
down_days = df_future[df_future["Pred_Class"] == 0].index

plt.scatter(up_days, df_future.loc[up_days, "Close"], marker="^", color="green", label="Predicted Up")
plt.scatter(down_days, df_future.loc[down_days, "Close"], marker="v", color="red", label="Predicted Down")

plt.title("AAPL Predictions for 2025")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()

sns.lineplot(x=df_future.index, y=df_future["Pred_Prob"], color="purple")
plt.axhline(0.5, color="gray", linestyle="--", alpha=0.7)
plt.title("Predicted Probability of Up Move (2025)")
plt.ylabel("P(Up)")
plt.show()
```

```python
df_future["Strategy_Return"] = df_future["Daily_Return"] * df_future["Pred_Class"].shift(1)
df_future["Cum_Market"] = (1 + df_future["Daily_Return"]).cumprod()
df_future["Cum_Strategy"] = (1 + df_future["Strategy_Return"].fillna(0)).cumprod()

plt.plot(df_future.index, df_future["Cum_Market"], label="Market", color="blue")
plt.plot(df_future.index, df_future["Cum_Strategy"], label="Strategy", color="orange")
plt.title("Cumulative Returns in 2025")
plt.legend()
plt.show()
```

```python
out_path = Path("../outputs") / f"{ticker}_2025_predictions.csv"
out_path.parent.mkdir(parents=True, exist_ok=True)

df_future_out = df_future[["Close","SMA_20","EMA_50","Daily_Return","Lag_1","Pred_Class","Pred_Prob"]].copy()
df_future_out.reset_index().rename(columns={"index":"Date"}).to_csv(out_path, index=False)

print(f"Saved predictions to: {out_path}")
```

```python
def predict_direction_for_date(date_str, df_future, model):
    date = pd.to_datetime(date_str)
    if date not in df_future.index:
        return f"No data available for {date_str}"

    features = list(model.feature_names_in_)
    row = df_future.loc[[date], features]  # keep DataFrame with names

    pred_class = model.predict(row)[0]
    pred_prob = model.predict_proba(row)[0, 1]

    direction = "Up" if pred_class == 1 else "Down"
    return {
        "Date": date_str,
        "Prediction": direction,
        "Probability of Up": round(pred_prob, 3)
    }
```

```python
result = predict_direction_for_date("2025-07-15", df_future, model)
print(result)
```
